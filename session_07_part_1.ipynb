#### Handling missing values and Errors

To support imputation and improve feature representation, we add three new features:
-  *price_per_sq_meter* — price normalized by total square footage
-  *living_ratio* — ratio of living space to total square footage
-  *kitchen_ratio* — ratio of kitchen space to total square footage

---

These features can improve data quality, support missing value handling, and help improve model performance.

```python
df['price_per_sq_meter'] = (df['last_price'] / df['total_area']).round(2)
```

```python
df.info()
```

```python
df["ceiling_height"] = df["ceiling_height"].apply(lambda item: item if 2.3 <= item <= 4.2 else None)
mean_ceiling_by_type = df.groupby("is_apartment")["ceiling_height"].transform("mean")
df["ceiling_height"] = df["ceiling_height"].fillna(mean_ceiling_by_type)
```

```python
# If the value in the floors_total column is missing (None or NaN),
# then replace it with the corresponding value from the floor column — assuming the property is located on the top floor.
df["floors_total"] = df["floors_total"].fillna(df["floor"])
```

```python
df.loc[df["studio"] == True, "kitchen_area"] = 0
mask = df['total_area'] < (df['living_area'] + df['kitchen_area'])
print(mask.sum())
```

```python
sum_parts = df["living_area"] + df["kitchen_area"]
correction_factor = df["total_area"] / sum_parts

df.loc[mask, "living_area"] *= correction_factor[mask]
df.loc[mask, "kitchen_area"] *= correction_factor[mask]
```

```python
df["living_ratio"] = df["living_area"] / df["total_area"]

median_ratio = df.groupby("is_apartment")["living_ratio"].transform("median")
df["living_area"] = df["living_area"].fillna(df["total_area"] * median_ratio)
```

```python
df["kitchen_ratio"] = df["kitchen_area"] / df["total_area"]
median_ratio = df.groupby("is_apartment")["kitchen_ratio"].transform("median")

df.loc[df["studio"] == True, "kitchen_area"] = 0
df["kitchen_area"] = df["kitchen_area"].fillna(df["total_area"] * median_ratio)
```

```python
from datetime import datetime

today = pd.to_datetime("today")
df["days_exposition"] = df["days_exposition"].fillna((today - df["first_day_exposition"]).dt.days)
```

```python
def categorize_floor(row):
  if row['floor'] == 0:
    return 'ground'
  elif row['floor'] == 1:
    return 'first'
  elif row['floor'] == row['floors_total']:
    return 'last'
  else:
    return 'middle'

df['floor_type'] = df.apply(categorize_floor, axis=1)
```

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], errors='coerce')
df['iso_day_of_week'] = df['first_day_exposition'].dt.isocalendar().day
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```
