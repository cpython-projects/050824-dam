```python
class Pizza:

    __pizzas = {
        'margarita': 'томаты, моцарелла, базилик',
        'pepperoni': 'томатный соус, моцарелла, пепперони',
        'hawaiian': 'томатный соус, моцарелла, ветчина, ананасы',
        'veggie': 'томатный соус, моцарелла, болгарский перец, лук, грибы, маслины',
        'meat_lovers': 'томатный соус, моцарелла, ветчина, пепперони, бекон, колбаса',
        'bbq_chicken': 'барбекю соус, курица, красный лук, моцарелла, кинза',
        'four_cheese': 'моцарелла, пармезан, горгонзола, рикотта',
        'seafood': 'томатный соус, моцарелла, креветки, мидии, кальмары',
        'mexican': 'томатный соус, моцарелла, острый перец халапеньо, кукуруза, фарш',
        'carbonara': 'сливочный соус, моцарелла, бекон, пармезан, яйцо'
    }

    def __init__(self, name):
        if name not in Pizza.__pizzas:
            raise ValueError('Unknown pizza name')
        self.__name = name

    @classmethod
    def get_pizza(cls, name):
        if name not in Pizza.__pizzas:
            raise ValueError('Unknown pizza name')

        return cls.__pizzas[name]

    @staticmethod
    def pizza_size(size: str = 'small'):
        if size == 'small':
            return 15
        if size == 'medium':
            return 20
        if size == 'large':
            return 25

    def __str__(self):
        return f'{self.__name}: {Pizza.__pizzas[self.__name]}'


x = Pizza('margarita')
print(x)

print(Pizza.pizza_size('medium'), 'cm')
print(Pizza.get_pizza('margarita'))
```

```python
import logging

logger = logging.getLogger('ich')
logger.setLevel(logging.DEBUG)
formatter = logging.Formatter('%(asctime)s %(name)-12s %(levelname)-8s %(message)s')

console = logging.StreamHandler()
console.setLevel(logging.DEBUG)
console.setFormatter(formatter)

filehandler = logging.FileHandler('data.log')
filehandler.setLevel(logging.INFO)
filehandler.setFormatter(formatter)

logger.addHandler(console)
logger.addHandler(filehandler)

class Box:
    def __init__(self, x, y, z):
        if not isinstance(x, int):
            logger.error('x must be an integer')
            raise TypeError('x must be an integer')

        if not isinstance(y, int):
            logger.error('y must be an integer')
            raise TypeError('y must be an integer')

        if not isinstance(z, int):
            logger.error('z must be an integer')
            raise TypeError('z must be an integer')

        self.x = x
        self.y = y
        self.z = z

        logger.debug('Box(%d, %d, %d)', self.x, self.y, self.z)
        logger.info('Box was created')


try:
    x = Box(1, 2, 3)
    y = Box('2', 3, '4')
except Exception as e:
    print(e)
```

```python
import datetime
import logging

logging.basicConfig(
    filename='data.log',
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
)


def log_args(func):
    def wrapper(*args, **kwargs):
        logging.info(f'{func.__name__} - {args}, {kwargs}')
        res = func(*args, **kwargs)
        logging.info(f'{func.__name__} - {res}')
        return res
    return wrapper


@log_args
def hello_user(name: str) -> str:
    return f"Hello {name}"

@log_args
def user_age(name: str, date_of_birth: str) -> str:
    date_of_birth = datetime.datetime.strptime(date_of_birth, "%d-%m-%Y")
    age = datetime.datetime.now().year - date_of_birth.year
    return f'{name} is {age} years old.'


print(hello_user('Jack'))
print(user_age('Jack', '01-01-2000'))
```
