```python
import json

class Student:
    def __init__(self, first_name, last_name):
        self.first_name = first_name.strip().title()
        self.last_name = last_name.strip().title()
        self.id = None
        
    
    def __str__(self):
        return f'{self.last_name} {self.first_name[0]}.'

    def __repr__(self):
        return json.dumps(self.__dict__)




x1 = Student('ivan', 'ivanov')
x2 = Student('petr', 'petrov')

students = [x1, x2]

for item in students:
    print(item.__repr__(), type(item.__repr__()))



tmp = '{"first_name": "Ivan", "last_name": "Ivanov"}'
tmp = json.loads(tmp)
print(tmp, type(tmp))

new_st = Student(**tmp)
print(new_st)



# import json

# tmp = json.dumps(students)

# print(students_str)

```

```python
def to_file(func):
    
    def wrapper(*args, **kwargs):
        res = func(*args, **kwargs)
        with open(file_name, 'a') as f:
            f.write(str(res) + '\n')
        return res
    return wrapper


@to_file
def greeting(name):
    return f"Hello, {name}"


@to_file
def hello(first_name, last_name):
    return f"Hello, {first_name} {last_name}"


print(greeting('John'))
print(hello('John', 'Doe'))
```

```python
import datetime
def to_file(file_name):
    def decorator(func):
        def wrapper(*args, **kwargs):
            print(datetime.datetime.now().strftime("%I:%M%p on %B %d, %Y"))
            res = func(*args, **kwargs)
            with open(file_name, 'a') as f:
                f.write(str(res) + '\n')
            print(datetime.datetime.now().strftime("%I:%M%p on %B %d, %Y"))
            return res
        return wrapper
    return decorator


@to_file('data.txt')
def greeting(name):
    return f"Hello, {name}"


@to_file('test.txt')
def hello(first_name, last_name):
    return f"Hello, {first_name} {last_name}"


print(greeting('John'))
print(hello('John', 'Doe'))
```

```python
def text_modifier(char):
    def decorator(func):
        def wrapper(*args, **kwargs):
            return f'{char}{func(*args, **kwargs)}{char}'
        return wrapper
    return decorator


def to_file(func):
    def wrapper(*args, **kwargs):
        res = func(*args, **kwargs)
        with open('test.txt', 'w') as f:
            f.write(f'{res}\n')
        return res
    return wrapper


@text_modifier('*')
@text_modifier('#')
@to_file
def func_1(text):
    return text


print(func_1('Oleh'))
```
